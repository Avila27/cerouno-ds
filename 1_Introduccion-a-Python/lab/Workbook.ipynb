```python
r = requests.get("https://www.nasdaq.com/symbol/aapl/historical")
```

```python
from bs4 import BeautifulSoup
```

```python
soup= BeautifulSoup(r.content)
```

```python
value_table = soup.find_all("table")[2]
```

```python
table_rows = value_table.find_all("tr")
```

```python
def clean_string(text):
    return text.replace("\r","").replace(" ", "").replace("\n","")
```

```python
output_table = []
for row in table_rows:
    output_table.append([clean_string(td.text) for td in row.find_all("td")])
```

```python
pd.DataFrame(output_table).dropna()
```
